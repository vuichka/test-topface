1. Рекламная кампания стартовала вчера с дневным бюджетом 40 $. Половина бюджета была израсходована к полудню, а 80% оставшегося бюджета было потрачено между полуднем и временем закрытия. Сколько долларов не было
потрачено?

Решается простыми вычислениями $$ 40 - (40 * 0.5) - (40 * 0.5 * 0.8)$$

```python
40 - (40 * 0.5) - (40 * 0.5 * 0.8)
```

**Ответ :** 4 доллара не было потрачено

2. Пять идентичных рекламных кампаний работали 24 дня по 6 часов в день, потрачено было 120 долларов. Сколько дней они работали бы на 216 долларов, если бы 9
одинаковых кампаний работали бы по 8 часов в день?

Сначала надо найти сколько тратится долларов в час.

$$ 5 * 24 * 6 * x = 120 $$

```python
120 / (5 * 24 * 6)
```

$ x = \dfrac{1}{6} $ - Тратили на рекламу долларов в час

Теперь найдем сколько бы они работали, если бы тратили в час столько же

$$ 9 * y * 8 * \dfrac{1}{6} = 216 $$

```python
216 / (9 * 8 * 1/6)
```

**Ответ :** 18 полных дней бы проработали рекламные компании.

3. Дизайнеры создали 200 рекламных баннеров для двух рекламных кампаний. 80 из них не использовались ни в кампании No1, ни в кампании No2, 60 использовались только в кампании No1. И для каждого рекламного баннера, который использовался в обеих кампаниях, приходится 3 баннера, которые использовались только в кампании No2. Сколько баннеров было использовано в обеих рекламных

| 1 и 2  | не 1 и 2  | 1 и не 2   | не 1 и не 2  |
|---|---|---|---|
| y  | x  | 60  | 80  |


По условию задачи знаем, что всего их 200. Значит $ x + y = 200 - (60 + 80) = 60$

$$ \begin{cases}
x = 3y\\
x + y = 60
\end{cases} $$



Решаем систему, получаем ответы $ x = 45\\y = 15 $

**Ответ :** 15

```python
def str_replacer(x):
    MONTHS = [
        'Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь', 'Июль', 'Август',
        'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь'
    ]

    m, y = x.strip().split()
    m = MONTHS.index(m) + 1
    return "-".join((str(m), y))


df['dt'] = df['dt'].apply(str_replacer)
```

1. За новые сделки менеджер получает 7 % от суммы, при условии, что статус оплаты «ОПЛАЧЕНО», а также имеется оригинал подписанного договора с клиентом (в рассматриваемом месяце).
1. За текущие сделки менеджер получает 5 % от суммы, если она больше 10 тыс., и 3 % от суммы, если меньше. При этом статус оплаты может быть любым, кроме «ПРОСРОЧЕНО», а также необходимо наличие оригинала подписанного договора с клиентом (в рассматриваемом месяце). Бонусы по сделкам, оригиналы для которых приходят позже рассматриваемого месяца, считаются остатком на следующий период, который выплачивается по мере прихода оригиналов.

Вычислите остаток каждого из менеджеров на 01.07.2021.

```python
df.status.unique()
```

```python
def conditions(s):
    if s['status'] == 'оплачено' and s['document'] == 'оригинал':
        return s['sum'] * 0.07
    if s.status in ['в работе', 'на подписании'] and s['document'] == 'оригинал':
        if s['sum'] > 10000:
            return s['sum'] * 0.05
        else:
            return s['sum'] * 0.03
```

```python
df['bonus'] = df.apply(conditions, axis=1)
```

```python
fig = plt.figure(figsize=(9, 6))

ax = sns.barplot(data=df[df['dt'] <= '2021-07-01'].groupby(
    'sale', as_index=False).agg({
        'bonus': 'sum'
    }).sort_values('bonus', ascending=False),
                 x='bonus',
                 y='sale',
                 color="#BBBBBB")
for i in range(2):
    ax.patches[i].set_color('#547CBD')
ax.bar_label(ax.containers[0], fontsize=8, color='#808080')

plt.suptitle('Amount of bonuses by salesmen', fontsize=26,\
             x=0.3 ,va='top', color='#547CBD', y=1)

sns.despine()
ax.spines['left'].set_color('lightgray')
ax.spines['bottom'].set_color('lightgray')
plt.xlabel('Amount of money', color='#616161')
plt.ylabel('')
```
